```python
#taking out top 3 cast members
def cast(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj) if isinstance(obj, str) else []:
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
# Now Calculating distance among each movie in a 34k vector space, we will calcualte cosine distance i.e angle b/w each vectors, not euclidean distance
# euclidian distance is not a good measure for higher dimensional data, " curse of Dimensionality "

# DISTANCE AND SIMILARITY IS INVERSELY PROPORTIONAL, i.e IF THE DISTANCE B/W TWO VECTORS IS MORE THEN THE SIMILARITY IS LESS 
```

```python
# cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix

# Convert to sparse matrix
sparse_vectors = csr_matrix(vectors)

# Function to compute sparse similarity
def compute_sparse_similarity(sparse_matrix, top_k=500):
    num_rows = sparse_matrix.shape[0]
    similarity_sparse = {}

    for i in range(num_rows):
        row_sim = cosine_similarity(sparse_matrix[i], sparse_matrix).flatten()
        top_indices = row_sim.argsort()[-top_k-1:-1]  # Get top k similar items
        similarity_sparse[i] = {idx: row_sim[idx] for idx in top_indices}

    return similarity_sparse

# Compute sparse similarity
similarity_sparse = compute_sparse_similarity(sparse_vectors, top_k=500)

print("Computed sparse cosine similarity successfully!")
```

```python
def similarity_dict_to_matrix(similarity_dict, num_movies, top_k=500):
    
    similarity_matrix = np.zeros((num_movies, top_k))  # Initialize empty matrix
    
    for i in range(num_movies):
        if i in similarity_dict:
            # Get top-K similar movies and their scores
            sorted_similarities = sorted(similarity_dict[i].items(), key=lambda x: x[1], reverse=True)[:top_k]
            
            # Extract only similarity scores (ignoring movie indices)
            similarity_matrix[i, :len(sorted_similarities)] = [score for _, score in sorted_similarities]
    
    return similarity_matrix

# Convert dictionary to similarity matrix (keeping top-20 closest movies)
num_movies = sparse_vectors.shape[0]  
similarity_matrix = similarity_dict_to_matrix(similarity_sparse, num_movies, top_k=500)

# Print matrix shape and sample row
print("Similarity Matrix Shape:", similarity_matrix.shape)
print("Example row (similarity scores of 50 closest movies):\n", similarity_matrix)
```

```python
import random

def recommend(movie, n=20):
    try:
        # Get movie index
        movie_index = new_df[new_df['title'] == movie].index[0]
            
        # Get similarity scores of the movie (list of tuples with index and score)
        similarity_scores = list(enumerate(similarity_matrix[movie_index])) 
        
        # Sort movies based on similarity scores
        sorted_similarities = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:251]
        
        # Get the indices of the top 250 similar movies (excluding itself)
        random.shuffle(sorted_similarities)  # Shuffle the top 250 movies

        # Select 'n' random movies from the top 250 similar movies
        selected_titles = [new_df['title'].iloc[i] for i, _ in sorted_similarities[:n]]
        
    except IndexError:
        print(f"❌ '{movie}' not found! Suggesting 20 random movies instead.")
        selected_titles = new_df['title'].sample(n).tolist()

    print("\n".join(selected_titles))
```
